```python
#CIFAR10 veriseti ile veri artirma 
#Load Libraries
from keras.datasets import cifar10
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from keras import backend as krs
import numpy as np
```

```python
#Verisetini test/train oluşturma
(X_train, Y_train), (X_test, Y_test)=cifar10.load_data() #Veri indirme

X_train=X_train.astype('float32')
X_test=X_test.astype('float32')

X_train/=255 #X_train=X_train/255
X_test/=255
```

```python
images=ImageDataGenerator()
images.fit(X_train)

for X_batch, y_batch in images.flow(X_train,Y_train, batch_size=4, seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(X_batch[i])
  pyplot.show()
  break
```

```python
#DÖNDÜRME
images=ImageDataGenerator(rotation_range=270)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```

```python
#Dik Kaydırma
images=ImageDataGenerator(height_shift_range=0.7)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```

```python
#Yatay Simetri
images=ImageDataGenerator(horizontal_flip=True)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```

```python
#Yatay Kaydırma
images=ImageDataGenerator(width_shift_range=0.3)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```

```python
#Kırpma
images=ImageDataGenerator(shear_range=0.8)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```

```python
#Yaklaştırma
images=ImageDataGenerator(zoom_range=0.5)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```

```python
#Dik Simetri
images=ImageDataGenerator(vertical_flip=True)
images.fit(X_train)

for x_batch, y_batch in images.flow(X_train, Y_train, batch_size=4,seed=443):
  for i in range(0,4):
    pyplot.subplot(220+1+i)
    pyplot.imshow(x_batch[i])
  pyplot.show()
  break
```
