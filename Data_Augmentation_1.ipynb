```python
#Load Libraries
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
#Data Augmentation Techniques
data_aug=ImageDataGenerator(rotation_range=45,
                            width_shift_range=0.4,
                            height_shift_range=0.4,
                            shear_range=0.3,
                            zoom_range=0.5,
                            horizontal_flip=True,
                            vertical_flip=True,
                            fill_mode='nearest')
```

```python
#Load İmage
image=load_img('image578.jpg')
x=img_to_array(image)
x = x.reshape((1,) + x.shape)
```

```python
#Bir resimden birden fazla üretmek için..
i=0
for img in data_aug.flow(x,batch_size=1,
                         save_to_dir='/gdrive/My Drive/DeepLearningNotes/Veri_Artirma/',
                         save_format='jpg'):
  i+=1
  if i>50:
    break
```
